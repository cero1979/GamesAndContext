# Reproducing context-dependent benefit-loss games

This notebook is a readable interface to the tested `context_games` package. The model and theorem checks live in `src/` and `tests/`; no result depends on hidden notebook state.

```python
from pathlib import Path
import platform
import matplotlib
import numpy as np
import pandas as pd
from context_games.benchmarks import BENCHMARKS
from context_games.experiments import (baseline_tables, convergence_diagnostics, exhaustive_dichotomy_audit, perturbation_audit, run_all, sensitivity_grid, sensitivity_grid_summary)

RESULTS = Path('results')
print({'python': platform.python_version(), 'numpy': np.__version__, 'pandas': pd.__version__, 'matplotlib': matplotlib.__version__})
```

## 1. Payoffs, classes, and equilibria

```python
tables = baseline_tables()
display(tables['payoff_table'])
display(tables['class_table'])
display(tables['pure_nash'])
```

## 2. Baseline dynamics and finite-horizon diagnostics

```python
display(tables['final_values'])
display(convergence_diagnostics())
```

The horizon check reports the one-step residual and the change relative to the previous horizon. It guards against reporting an arbitrary finite iterate as a converged result.

## 3. Deterministic design over initial conditions and intensities

```python
grid = sensitivity_grid()
display(sensitivity_grid_summary(grid))
grid.head()
```

The design points are not random observations. Ranges are therefore reported instead of standard errors or inferential summaries.

## 4. Payoff-perturbation audit

```python
perturbations = perturbation_audit()
display(perturbations)
```

For the business benchmark, the class map is locally robust but the original two-equilibrium set is not: continuous perturbations break the exact actor indifference with probability one.

## 5. Exhaustive finite audit of the configuration theorem

```python
display(exhaustive_dichotomy_audit())
```

All $4^8=65{,}536$ games with payoff coordinates in $\{-2,-1,1,2\}$ are checked. This is a finite computational audit, not a replacement for the proof.

## 6. Rebuild archived tables and figures
